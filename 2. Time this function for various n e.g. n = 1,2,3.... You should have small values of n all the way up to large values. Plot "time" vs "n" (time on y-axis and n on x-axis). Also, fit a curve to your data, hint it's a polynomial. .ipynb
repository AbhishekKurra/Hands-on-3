```python
import time
import numpy as np
import matplotlib.pyplot as plt

def f(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x += 1
    return x

n_values = np.arange(1, 1000, 50)
times = []

for n in n_values:
    start_time = time.time()
    f(n)
    times.append(time.time() - start_time)


coeffs = np.polyfit(n_values, times, 3)
poly = np.poly1d(coeffs)


plt.scatter(n_values, times, label="Measured Time")
plt.plot(n_values, poly(n_values), label=f"Fitted Curve: {poly}", color='green')
plt.xlabel("n")
plt.ylabel("Time (seconds)")
plt.legend()
plt.title("Runtime of f(n)")
plt.show()
```
